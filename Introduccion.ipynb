Una serie de tiempo es una collección de observaciones o datos los cuales dependende del tiempo. Podemos encontrar diferentes en diferentes campos de estudio como lo puede ser la física, biología, economía, marketing, ingeniería. 

<b>Economía:</b> En el campo de la economía las series de tiempo presentan una herramienta fundamental y son utiizadas comunmente en precios de acciones, exportaciones, ventas e ingresos.

<b>Física </b> No es una sorpresa el hecho que la ciencia utilice esta herramienta, un ejemplo claro de una serie de tiempo es la expansión del universo. Ya que sabemos que con el paso del tiempo este se expande. Igualmente estas son utilizadas en diferentes subdisciplinas como la meteorología, ciencas marinas y geofísica. 

A continuación se presenta un ejemplo de una serie de tiempo.

```python
#Importamos librerías
import pandas as pd
import matplotlib.pyplot as plt
#Guardamos datos
with open("cambio_climatico.txt") as f:
    lines = f.readlines()
    x = [line.split()[0] for line in lines]
    y = [line.split()[1] for line in lines] 
#Eliminamos primeros terminos
x.remove("Year")
y.remove("No_Smoothing")
#Convertimos cadena a float y entero
for i in range(0,len(x)):
    x[i] = int(x[i])
    y[i] = float(y[i])

#Graficamos
plt.plot(x,y)
plt.xlim(min(x)-2, max(x)+2)
plt.ylim(-0.5, 0.5)
plt.xticks(x[::10],  rotation='vertical');
plt.xlabel("Año")
plt.ylabel("Incremento de temperatura en C$^\circ$")
plt.ylim((min(y)-0.1, max(y)+0.1));
```

Datos obtenidos de: <a href="https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.txt">Land-Ocean Temperature Index (C)</a>

Como se puede apreciar en la figura anterior tenemos que la temperatura parece tener un comporatmiento ascendente con el paso de los años. Existen diferentes tipos de series de tiempo en las que se encuentran <b> continuas </b> y <b> discretas</b>. La figura anterior representa una serie de distribución <b>continua</b> y su característica principal es que las observaciones son realizadas en un periodo continuo de tiempo. Se dice que tenemos una serie de tiempo <b>discreta</b> cuando las mediciones son realizadas en tiempos específicos, normalmente estas mediciones son realizadas en tiempos perdiódicos. Una vez cada hora, cada cinco días, cada año, etc. El término "discreto" es utilziado para estas series de tiempo aunque dichas medidas sean realizadas en un periodo de tiempo continuo.
